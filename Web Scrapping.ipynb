```python
#importing basic libraries required
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#the given url
URL = "https://www.amazon.in/s?rh=n%3A6612025031&fs=true&ref=lp_6612025031_sar"
```

```python
#adding the user agent
HEADERS=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36','Accept-Language':'en-US, en; q=0.5'})

#http request
web=requests.get(URL,headers=HEADERS)

#Creating soup object to contain all the data
soup=BeautifulSoup(web.content,'html.parser')

#Fetching the links using tag object
links=soup.find_all('a',attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})

#defining a list to store the links
link_list=[]

#Extracting links and appending in the list
for i in links:
    link_list.append(i.get('href'))
    
#defining a dictionary to store the details separately
dict={'Name':[],'Price':[],'Rating':[],'SellerName':[]}
    
#Generating the complete link and the product details
for i in link_list:
    com_web=requests.get("https://amazon.in/"+i, headers=HEADERS)
    com_soup=BeautifulSoup(com_web.content,"html.parser")
    
    #Calling the functions to get the details of a product
    dict['Name'].append(get_Name(com_soup))
    dict['Price'].append(get_Price(com_soup))
    dict['Rating'].append(get_Rating(com_soup))
    dict['SellerName'].append(get_SellerName(com_soup))
    
data=pd.DataFrame.from_dict(dict)
data.to_csv("Scraped_data.csv",header=True,index=False)
```

```python
#different functions to extract different product details

#function to extract name of the product
def get_Name(soup):
    try:
        name=soup.find("span",attrs={'id':'productTitle'}).text.strip() #.text is used to extract the string part and .strip() is used to eliminate the excess space
    except AttributeError:
        name=""
        
    return name

#function to extract price of the product
def get_Price(soup):
    try:
        price=soup.find("span",attrs={'class':'a-price-whole'}).text
    except AttributeError:
        price=""
    return price

#function to extract rating
def get_Rating(soup):
    try:
        rating=soup.find("i",attrs={'class':'a-icon a-icon-star a-star-4'}).text.strip()
        #rating2=soup.find("i",attrs={'class':'a-icon a-icon-star-small a-star-small-4-5 aok-align-bottom'}).text.strip()
    except AttributeError:
        rating1=""
        #rating2=''
    return rating

#function to extract seller name
def get_SellerName(soup):
    try:
        sellername=soup.find("div",attrs={'id':'merchant-info'}).text.strip() 
    except AttributeError:
        sellername=""       
    return sellername
```
